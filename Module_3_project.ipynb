```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://btech-data-analytics.github.io/bridgerland-technical-college/home.html")
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
books = soup.find_all("div", class_="product-card")

print(len(books))
```

```python
titles = []
authors = []
topics = []
prices = []
pages = []
```

```python
for book in books:

    title = book.find("h3").text
    author = book.find_all("p")[0].text.replace("Author: ", "")
    topic = book.find_all("p")[1].text.replace("Topic: ", "")
    price = float(book.find_all("p")[2].text.replace("Price: $", ""))
    page = int(book.find_all("p")[3].text.replace("Pages: ", ""))

    titles.append(title)
    authors.append(author)
    topics.append(topic)
    prices.append(price)
    pages.append(page)
```

```python
df = pd.DataFrame({
    "Title": titles,
    "Author": authors,
    "Topic": topics,
    "Price": prices,
    "Pages": pages
})

df.head()
```
