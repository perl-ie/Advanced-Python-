```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("housing.csv")
```

```python
#1
sns.scatterplot(x="Landsize", y="Price", hue="Type", data=df)

plt.show()
```

```python
#2
House
```

```python
#3
df.groupby("YearBuilt")["Price"].mean().plot()

plt.show()
```

```python
#4
As the Property becomes newer, The Average price increases
```

```python
#5
df.groupby("Regionname")["Price"].mean().plot(kind="bar")

plt.show()
```

```python
#6
Southern Metropolitan
```

```python
#7
sns.histplot(df["Price"])

plt.show()
```
