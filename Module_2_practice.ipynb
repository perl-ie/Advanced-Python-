```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("companies.csv")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="state", data=df)

plt.xticks(rotation=90)
plt.show()
```

```python
industry = df.groupby("Industry")["job_openings"].sum()

industry.plot(kind="bar", figsize=(10,5))

plt.show()
```

```python
growth = df.groupby("founded")["growth_percentage"].mean()

growth.plot(figsize=(10,5))

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="employees", y="estimated_revenues", data=df)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["founded"])

plt.show()
```
